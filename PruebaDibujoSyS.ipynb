```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*60* np.pi, 60)
y = (120*np.cos(x))

def my_ceropen(x,ymin = 0, ymax = 15):
  xmax = max(x) # x.max()
  xmin = min(x) # x.min()
  m = (ymax - ymin)/(xmax - xmin)
  c = ymin - m*xmin
  yv = m*x + c
  return yv
yv = my_ceropen(y,ymin=0,ymax=5)

plt.stem(x, yv, use_line_collection=True)
plt.show()
```
